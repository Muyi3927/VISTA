```python
import torch
import torch.nn as nn
import numpy as np
from tqdm import tqdm
import torch.nn.functional as F
from dift_sd import SDFeaturizer
from PIL import Image
from torchvision.transforms import PILToTensor
import matplotlib.pyplot as plt
```

```python
def visualize_correspondence_lines(target_img, ref_img, match_dict, step=8, max_lines=1000, figsize=(10, 10)):
    """
    target_img, ref_img: ndarray 或 PIL.Image
    match_dict: {(y, x): (y', x')}
    step: 跳步间隔（建议与计算时一致）
    max_lines: 最多画多少根线
    """
    if hasattr(target_img, 'numpy'):  # 兼容 torch.Tensor 输入
        target_img = target_img.permute(1, 2, 0).cpu().numpy()
    if hasattr(ref_img, 'numpy'):
        ref_img = ref_img.permute(1, 2, 0).cpu().numpy()

    # 图像转为 numpy 格式（H, W, C）
    target_img = np.array(target_img)
    ref_img = np.array(ref_img)

    # 拼接图像（水平拼）
    combined_img = np.concatenate([target_img, ref_img], axis=1)
    h, w, _ = target_img.shape
    offset = w  # 第二张图的 x 偏移量

    # 可视化
    plt.figure(figsize=figsize)
    plt.imshow(combined_img.astype(np.uint8))
    plt.axis('off')

    # 抽样画线
    keys = list(match_dict.keys())[::step]
    if len(keys) > max_lines:
        keys = keys[:max_lines]

    for (y1, x1) in keys:
        y2, x2 = match_dict[(y1, x1)]
        plt.plot([x1, x2 + offset], [y1, y2], 'k-', linewidth=0.5)
        plt.scatter([x1], [y1], c='red', s=5)
        plt.scatter([x2 + offset], [y2], c='red', s=5)

    plt.tight_layout()
    plt.show()


def compute_dense_correspondence(target_ft, ref_ft, img_size=512, step=8):
    """
    target_ft, ref_ft: torch.Tensor, shape [C, H, W]
    return: dict[(ty, tx)] = (ry, rx)
    """
    device = target_ft.device

    # 插值到一致大小（加 batch 维度）
    target_ft_up = F.interpolate(target_ft, size=(img_size, img_size), mode='bilinear', align_corners=False)[0]
    ref_ft_up = F.interpolate(ref_ft, size=(img_size, img_size), mode='bilinear', align_corners=False)[0]

    del target_ft, ref_ft
    torch.cuda.empty_cache()

    # 归一化
    target_ft_up = F.normalize(target_ft_up, dim=0)  # [C, H, W]
    ref_ft_up = F.normalize(ref_ft_up, dim=0)

    C, H, W = target_ft_up.shape
    match_dict = {}

    # 展平参考特征为 [C, H*W]
    ref_vecs = ref_ft_up.view(C, -1)  # [C, H*W]

    for y in tqdm(range(5, H, step), desc="Matching"):
        for x in range(5, W, step):
            tgt_vec = target_ft_up[:, y, x].view(1, C)  # [1, C]
            sim = torch.matmul(tgt_vec, ref_vecs)  # [1, H*W]
            idx = sim.argmax().item()
            ry, rx = idx // W, idx % W
            match_dict[(y, x)] = (ry, rx)

    return match_dict
```

```python
torch.cuda.set_device(1)
```

```python


dift = SDFeaturizer()

ft = []
imglist = []
tar_img_path = "/home/hjh/repository/VISTA/dataset/buildings/2.jpg"
ref_img_path = "/home/hjh/repository/VISTA/dataset/buildings/1.jpg"
prompt = f'a photo of a buildings'
filelist = [tar_img_path, ref_img_path]
img_size = 768
ensemble_size = 8
```

```python
torch.cuda.empty_cache()
for filename in filelist:
    img = Image.open(filename).convert('RGB')
    img = img.resize((img_size, img_size))
    imglist.append(img)
    img_tensor = (PILToTensor()(img) / 255.0 - 0.5) * 2
    ft.append(dift.forward(img_tensor,
                           prompt=prompt,
                           ensemble_size=ensemble_size))
```

```python

torch.cuda.empty_cache()
```

```python
match_dict = compute_dense_correspondence(ft[0], ft[1], img_size=768, step=32)
```

```python
visualize_correspondence_lines(imglist[0], imglist[1], match_dict, step=8, max_lines=600, figsize=(10, 10))
```
